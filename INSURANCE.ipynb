```python
df1 = df
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = df1.drop(['charges'], axis = 1)
y = df1['charges']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 7)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
print('ACCURACY OF MODEL IS :', lr.score(x_test, y_test))
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
df2 = df
```

```python
from sklearn.ensemble import RandomForestRegressor as rfr
x = df2.drop(['charges'], axis = 1)
y = df2['charges']
```

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(x, y, test_size = 0.3, random_state = 7)
```

```python
regressor = rfr(n_estimators = 100, random_state = 10)
```

```python
regressor.fit(x_train1, y_train1)
```

```python
y_pred = regressor.predict(x_test1)
```

```python
df_f = pd.DataFrame({'ACTUAL':y_test1, 'PREDICTED':y_pred})
```

```python
df_f.head()
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test1, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test1, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test1, y_pred)))
```

```python
mae = abs(y_test1 - y_pred)
mae_f = round(np.mean(mae),2)

mape = 100 * (mae/y_test)           #MEAN ABSOLUTE PERCENTAGE ERROR

accuracy = 100 - np.mean(mape)

print('ACCURACY OF RANDOM FOREST REGRESSOR IS :', accuracy)
```

```python
#AS WE CAN SEE OUR LINEAR REGRESSION MODEL PERFORM WELL AS COMPARED TO THE RANDOM FOREST REGRESSOR
```
